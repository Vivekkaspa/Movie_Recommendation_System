```python
v= df2['vote_average'].mean()
v
```

```python
b= df2['vote_count'].quantile(0.9)
b
```

```python
# now we can filter out movies that qualify for the chart
q_movies=df2.copy().loc[df2['vote_count']>=b]
q_movies.shape
```

```python
def weighted_movie(x,b=b,v=v):
    p=x['vote_count']
    h=x['vote_average']
    #calculation based on IMDB formula
    
    return (p/(p+b)*h) + (b/(p+b)*v)
```

```python
#define a new feature 'score' and calculate its value with weighted_rating()

q_movies['score']= q_movies.apply(weighted_movie,axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)

#Print the top 15 movies
q_movies[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

## Content Based Filtering

```python
# In this recommender system the content of the movie (overview, cast, crew, keyword, tagline etc) is used to find its similarity with other movies.
#Then the movies that are most likely to be similar are recommended.
```

```python
df2['overview'].head(5)
```

```python
from ast import literal_eval

features = ["cast", "crew", "keywords", "genres"]

for feature in features:
    df2[feature] = df2[feature].apply(literal_eval)

df2[features].head(10)
```

```python
def get_director(x):
    for i in x:
        if i["job"] == "Director":
            return i["name"]
    return np.nan
```

```python
def get_list(x):
    if isinstance(x, list):
        names = [i["name"] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
df2["director"] = df2["crew"].apply(get_director)

features = ["cast", "keywords", "genres"]
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head()
```

## Data Cleaning & Pre-Processing

```python
# Removal of spaces
def clean_data(row):
    if isinstance(row, list):
        return [str.lower(i.replace(" ", "")) for i in row]
    else:
        if isinstance(row, str):
            return str.lower(row.replace(" ", ""))
        else:
            return ""

features = ['cast', 'keywords', 'director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

- Our movie recommendation engine works by suggesting movies to the user based on the metadata information. The similarity between the movies is calculated and then used to make recommendations. For that, our text data should be preprocessed and converted into a vectorizer using the CountVectorizer. 

```python
# create a metadata containing all of the metadata information extracted to input into the vectorizer
def create_metadata(features):
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + features['director'] + ' ' + ' '.join(features['genres'])


df2["metadata"] = df2.apply(create_metadata, axis=1)
print(df2["metadata"].head())
```

## Usage of Cosine-similarity score

```python
#resetting the indices of our dataframe

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_vectorizer = CountVectorizer(stop_words="english")
count_matrix = count_vectorizer.fit_transform(df2["metadata"])

print(count_matrix.shape)

cosine_sim2 = cosine_similarity(count_matrix, count_matrix) 
print(cosine_sim2.shape)

df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title'])
```
